```
import numpy as np
from keras.applications.inception_v3 import InceptionV3, preprocess_input
from keras.preprocessing import image
from keras.models import Model
from keras.layers import Input, Dense, Dropout
import os
```

```
image_shape = (299, 299, 3)
```

```
featureExtractor = InceptionV3(
    include_top=False, 
    weights='imagenet',
    pooling='avg', 
    input_shape=image_shape, 
    classes=1000)
featureExtractor.compile('sgd', 'mse')
```

```
datasetPath = '../datasets/coco/'
vectorsPath = 'data/inception_vectors/' 
```

```
imgpaths = [os.path.join(root, f) 
            for root, _, files in os.walk(datasetPath) 
            for f in files if f.endswith('.jpg')]
```

```
def extractAndSaveVectorsForBatch(batchImgPaths):
    batchSize = len(batchImgPaths)
    batch = np.zeros((batchSize,) + image_shape, dtype=np.float)
    
    for i in range(0, batchSize):
        img = image.img_to_array(image.load_img(batchImgPaths[i], target_size=image_shape))
        batch[i] = preprocess_input(img)
        
    batchVectors = featureExtractor.predict_on_batch(batch).astype(np.float16)
    for i in range(0, batchSize):
        np.save(batchImgPaths[i].replace(datasetPath, vectorsPath) + '.npy', batchVectors[i])
```

```
batchSize = 64
printEvery = 8
imgCount = len(imgpaths)
for i in range(0, imgCount, batchSize):
    extractAndSaveVectorsForBatch(imgpaths[i: i+batchSize])
    if not i % printEvery*batchSize:
        print('extracted vectors for %s images of %s' % (min(i+batchSize, imgCount), imgCount))
print('extracted vectors for every image')
```

```
imgVectors = []
for imgpath in imgpaths:
    vectorpath = imgpath.replace(datasetPath, vectorsPath) + '.npy'
    imgVectors.append(np.load(vectorpath))
    
inceptionVectors = np.array(imgVectors)
print('loaded all inception image vectors')
```

```
imgVectors = []
for imgpath in imgpaths:
    vectorpath = imgpath.replace(datasetPath, vectorsPath.replace('inception', 'resnet')) + '.npy'
    imgVectors.append(np.load(vectorpath))
    
resnetVectors = np.array(imgVectors)
print('loaded all resnet image vectors')
```

```
def getAugmentModel(input_size, output_size, embedding_size):
    inp = Input((input_size,))
    enc1 = Dense(embedding_size*4, activation='relu')(inp)
    enc2 = Dense(embedding_size*2, activation='relu')(enc1)
    bottleneck = Dense(embedding_size, activation='relu')(enc2)
    d = Dropout(0.3)(bottleneck)
    dec1 = Dense(embedding_size*2, activation='relu')(d)
    dec2 = Dense(embedding_size*4, activation='relu')(dec1)
    out = Dense(output_size, activation='relu')(dec2)
    model = Model(inputs=inp, outputs=out)
    model.compile('adam', 'mse')
    return model
```

```
model = getAugmentModel(
    input_size=resnetVectors.shape[-1], 
    output_size=inceptionVectors.shape[-1], 
    embedding_size=1024)
```

```
model.fit(inceptionVectors, resnetVectors, batch_size=64, epochs=100, validation_split=0.05)
```

```
def getEncoderModel(model):
    encoder = Model(inputs=model.inputs[0], outputs=model.layers[3].output)
    encoder.compile('adam', 'mse')
    return encoder
```

```
encoder = getEncoderModel(model)
```

```
embededVectors = encoder.predict(inceptionVectors, verbose=1)
```

```
from tsne import bh_sne
import cv2
```

```
embededVectors2d = bh_sne(embededVectors.astype(np.float))
```

```
canvasWidth = 6000
canvas = np.zeros((canvasWidth, canvasWidth, 3), dtype=np.uint8)
imgWidth = round((embededVectors2d.max() - embededVectors2d.min())*canvasWidth/10000)
imgWidth -= int(imgWidth % 2)
```

```
imgWidth = int(imgWidth)
```

```
minx = embededVectors2d[:, 0].min()
miny = embededVectors2d[:, 1].min()

embededVectors2d[:, 0] -= minx
embededVectors2d[:, 1] -= miny

maxx = embededVectors2d[:, 0].max()
maxy = embededVectors2d[:, 1].max()

canvasWidth = canvas.shape[1]
for i, imgpath in enumerate(imgpaths):
    imx = int((embededVectors2d[i, 0]/maxx * (canvasWidth-imgWidth)) + imgWidth/2)
    imy = int((embededVectors2d[i, 1]/maxy * (canvasWidth-imgWidth)) + imgWidth/2)
    scaledimage = cv2.resize(cv2.imread(imgpath), (imgWidth, imgWidth))
    canvas[imx - imgWidth/2: imx+imgWidth/2, imy-imgWidth/2: imy+imgWidth/2] = scaledimage
    if not i % 100:
        print('added % images on canvas ' % i)
```

```
cv2.imwrite('tsneCanvasEmbeded.png', canvas)
```
